```python
%%cuda -c "-I /does/not/exist -arch sm_75"
#include <stdio.h>
#include <cuda_runtime.h> // Include for cudaDeviceSynchronize, cudaSuccess, cudaGetErrorString, cudaPeekAtLastError, cudaDeviceReset

__global__ void kernel() {
    printf("Rabia Arif!\n");
}

int main() {
    kernel<<<1, 1>>>();

    // Error checking for kernel launch
    cudaError_t err = cudaDeviceSynchronize();
    if (err != cudaSuccess) {
        fprintf(stderr, "CUDA error: %s\n", cudaGetErrorString(err));
        exit(EXIT_FAILURE);
    }

    cudaDeviceReset(); // Reset the device when done
    return 0;
}
```
