```python
# fixing driver mismatch issue
%%cuda -c "-I /does/not/exist -arch sm_75"
#include <stdio.h>

__global__ void kernel() {
    printf("Hello World!\n");
}

int main() {
    kernel<<<1, 1>>>();
    cudaDeviceReset();
    return 0;
}
if(cudaDeviceSynchronize() != cudaSuccess{
    fprint(stderr,"cuda error: %s\n",cudaGetErrorString(cudaPeekAtLastError)));
    exit(EXIT_FAILURE);
}
return 0;}
```
